```python
import yaml

data_yaml_path = 'Dataset\\License-Plate-Data\\data.yaml'

# Read the existing data.yaml file
with open(data_yaml_path, 'r') as file:
    data = yaml.safe_load(file)

# Update the paths
# Assuming 'train', 'val', 'test' directories are directly under the base path
data['path'] = 'Dataset\\License-Plate-Data' # Base path for the dataset
data['train'] = 'train' # Assuming 'train' folder is directly inside the base path
data['val'] = 'test' # Assuming 'val' folder is not explicitly present, using 'test' as val set in this case, or if data.yaml uses 'test' for both
data['test'] = 'test' # Assuming 'test' folder is directly inside the base path

# Save the updated data.yaml file
with open(data_yaml_path, 'w') as file:
    yaml.dump(data, file, default_flow_style=False)

print(f"Updated {data_yaml_path} with new paths:")
print(yaml.dump(data, default_flow_style=False))
```

```python
from ultralytics import YOLO

# Load pretrained YOLOv8n
model = YOLO("yolov8n.pt")

# Train model with updated data path
model.train(
    data="Dataset\\License-Plate-Data\\data.yaml",
    epochs=50,
    imgsz=640
)
```
